```python
# Packages used
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import seaborn as sns
```

```python
# Original data files
cleaned_data = 'df5.csv'

# Read the data and check the results
df = pd.read_csv(cleaned_data)
df.head()
df.shape
```

```python
df['price_per_room'] = df.price/df.number_bedrooms
df['price_per_m2'] = df.price/df.area
df.dtypes
```

```python
df.subtype_property.value_counts()
```

```python
df_house = df[df['subtype_property'] == 'house']
df_villa = df[df['subtype_property']== 'villa']
df_appartment = df[df['subtype_property']== 'apartment']
print(df_house.shape)
print(df_villa.shape)
print(df_appartment.shape)
```

```python
df['price_per_m2'].plot.hist()
```

```python
df_house['price_per_m2'].plot.hist()
```

```python
print(df.subtype_property.value_counts())
```

```python
#Group by zone for Original dataset
price_per_room = df.groupby('Zone')['price_per_room'].mean()

df = pd.DataFrame(price_per_room[0:3])
df.head()
ax = df.plot.bar(rot=0)
```

```python
price_per_room = df.groupby('Zone')['price_per_room'].median()

df = pd.DataFrame(price_per_room[0:-1])
df.head()
ax = df.plot.bar(rot=0)
```

```python
#Group by zone for House dataset
price_per_room = df_house.groupby('Province')['price_per_room'].mean()

df = pd.DataFrame(price_per_room[0:11])
df.head()
ax = df.plot.bar(rot=40)
```

```python
price_per_room = df_house.groupby('Province')['price_per_room'].median()

df = pd.DataFrame(price_per_room[0:11])
df.head()
ax = df.plot.bar(rot=40)
```
